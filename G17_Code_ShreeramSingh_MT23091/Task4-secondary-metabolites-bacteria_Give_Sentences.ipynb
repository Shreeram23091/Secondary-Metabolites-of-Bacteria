```r
# Initialize an empty data frame to store the results
result_df <- data.frame(Chemicals = character(0), ChemicalsID = character(0), PMID = character(0), PMCID = character(0), Sentences = character(0), stringsAsFactors = FALSE)

# Read the CSV file containing pmcid, chemicals, chemicalsid, and pmid
pmcid_chemical_data <- read.csv("/content/Species_RemovedPMD.csv", header = TRUE)



# Loop through each row in the data frame
for (i in 1:nrow(pmcid_chemical_data)) {
  pmcid <- pmcid_chemical_data[i, "PMCID"]
  species <- pmcid_chemical_data[i, "Chemicals"]
  species_id <- pmcid_chemical_data[i, "ChemicalsID"]
  pmid <- pmcid_chemical_data[i, "PMID"]

  # Print progress
  print(paste("Chemicals:", species, ", PMCID:", pmcid))

  # Try to extract sentences; handle errors if any
  tryCatch(
    {
      # Call the Give_sentences_pmc() function to extract sentences
      sentences_matrix <- Give_Sentences_PMC(pmcid, species)

      # Flatten the matrix of sentences into a one-dimensional vector
      sentences_vector <- as.vector(sentences_matrix)

      # Create a data frame with one row containing chemicals, chemicalsid, pmid, pmcid, and flattened sentences
      result_row <- data.frame(Chemicals = species,
                               ChemicalsID = species_id,
                               PMID = pmid,
                               PMCID = pmcid,
                               Sentences = paste(sentences_vector, collapse = "; "))  # Concatenate sentences into one string

      # Add the result row to the result data frame
      result_df <- rbind(result_df, result_row)
    },
    error = function(e) {
      # Print error message
      print(paste("Error occurred while processing PMCID", pmcid, ":", conditionMessage(e)))

      # Create a data frame row with null values for sentences
      result_row <- data.frame(Chemicals = species,
                               ChemicalsID = species_id,
                               PMID = pmid,
                               PMCID = pmcid,
                               Sentences = NA)

      # Add the result row to the result data frame
      result_df <- rbind(result_df, result_row)
    }
  )
}

# Print the resulting data frame
print(result_df)

# Save the result data frame to a new CSV file
write.csv(result_df, "Species_Sentences.csv", row.names = FALSE)
```
