```python
import pandas as pd
df = pd.read_csv("/content/Task3_Secondary_Metabolites_Bacteria._Species_pubmed.csv")
species_list = [species.strip("[]'") for species in df['Species']]

print(species_list)
print(len(species_list))
```

# Change the organism with specieslist

```python
import requests
import csv
import gzip
import io

def download_and_save_tsv(url, file_path):
    try:
        response = requests.get(url)
        print(f"Status Code: {response.status_code}")
        if response.status_code == 200:
            content = response.content
            if content.startswith(b'\x1f\x8b'):
                content = gzip.decompress(content)
            with open(file_path, 'wb') as file:
                file.write(content)
            return True
        else:
            print(f"Failed to download the file: HTTP {response.status_code}")
            return False
    except Exception as e:
        print(f"An error occurred: {e}")
        return False

def append_tsv_to_csv(tsv_file_path, csv_file_path):
    try:
        with open(tsv_file_path, 'r') as tsvfile:
            tsv_reader = csv.reader(tsvfile, delimiter='\t')
            tsv_data = list(tsv_reader)

        with open(csv_file_path, 'a', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            if csvfile.tell() == 0:
                header = tsv_data[0]
                csv_writer.writerow(header)
            for row in tsv_data[1:]:
                csv_writer.writerow(row)

        return True
    except Exception as e:
        print(f"An error occurred while appending to CSV: {e}")
        return False

organisms = species_list   #only taking 10 for now

csv_file_path = 'merged_data.csv'


for organism in organisms:

    url = f"https://rest.uniprot.org/uniprotkb/stream?compressed=true&fields=accession%2Cid%2Cgene_names%2Cxref_geneid%2Clength%2Clit_pubmed_id%2Cprotein_name%2Corganism_name%2Ccc_function&format=tsv&query=({organism}))&size=500"


    tsv_file_path = f'{organism.replace(" ", "_")}_data.tsv'

    if download_and_save_tsv(url, tsv_file_path):
        print(f"TSV file saved for {organism}.")
        if append_tsv_to_csv(tsv_file_path, csv_file_path):
            print(f"TSV data appended to CSV for {organism}.")
        else:
            print(f"Failed to append TSV data to CSV for {organism}.")
    else:
        print(f"Failed to save TSV file for {organism}.")

print("Process completed.")
```
