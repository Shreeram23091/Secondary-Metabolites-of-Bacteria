```r
# Install and load necessary packages
install.packages(c("pubmed.mineR", "dplyr"))
library(pubmed.mineR)
library(dplyr)

# Read the original CSV file
original_data <- read.csv("Marburg.csv")

# Function to fetch PubTator data
fetch_pubtator_data <- function(pmid) {
  tryCatch({
    pubtator_result <- pubtator_function(pmid)

    # Replace NULL values with NULL
    genes <- if (is.null(pubtator_result$Genes) || !is.character(pubtator_result$Genes)) "NULL" else
      paste(pubtator_result$Genes, collapse=', ')

    diseases <- if (is.null(pubtator_result$Diseases) || !is.character(pubtator_result$Diseases)) "NULL" else
      paste(pubtator_result$Diseases, collapse=', ')

    mutations <- if (is.null(pubtator_result$Mutations) || !is.character(pubtator_result$Mutations)) "NULL" else
      paste(pubtator_result$Mutations, collapse=', ')

    chemicals <- if (is.null(pubtator_result$Chemicals) || !is.character(pubtator_result$Chemicals)) "NULL" else
      paste(pubtator_result$Chemicals, collapse=', ')

    species <- if (is.null(pubtator_result$Species) || !is.character(pubtator_result$Species)) "NULL" else
      paste(pubtator_result$Species, collapse=', ')

    return(data.frame(PMID = pmid, Genes = genes, Diseases = diseases,
                      Mutations = mutations, Chemicals = chemicals, Species = species))
  }, error = function(e) {
    cat("Error fetching PubTator data for PMID:", pmid, "\n")
    return(data.frame(PMID = pmid, Genes = "NULL", Diseases = "NULL",
                      Mutations = "NULL", Chemicals = "NULL", Species = "NULL"))
  })
}

# Apply the function to each PMID in the original data in the original order
pubtator_data_list <- lapply(original_data$PMID, fetch_pubtator_data)
pubtator_data_df <- do.call(rbind, pubtator_data_list)

# Merge the original data with the PubTator data
updated_data <- merge(original_data, pubtator_data_df, by = "PMID", all.x = TRUE)

# Reorder columns to place Genes, Diseases, Mutations, Chemicals, Species after PMID
updated_data <- updated_data[, c("PMID", "Genes", "Diseases", "Mutations", "Chemicals", "Species", setdiff(names(updated_data), c("PMID", "Genes", "Diseases", "Mutations", "Chemicals", "Species")))]

# Sort the data frame based on the order of PMIDs in the original CSV file
updated_data <- updated_data[match(original_data$PMID, updated_data$PMID),]

# Save the updated data to a new CSV file
write.csv(updated_data, "updated_data.csv", row.names = FALSE)

# Print a message indicating success
cat("Updated data saved to 'updated_data.csv'\n")
```
